```python
country = []
citizenship = driver.find_elements_by_xpath("//div[@class='countryOfCitizenship']")
for i in citizenship:
    country.append(i.text)
len(country)
```

A python program which searches all the product under a particular product vertical
from www.amazon.in. The product verticals to be searched will be taken as input from user.
For e.g. If user input is ‘guitar’. Then search for guitars.

```python
driver = webdriver.Chrome("chromedriver.exe")
```

```python
urls = "https://www.amazon.in/"
driver.get(urls)
```

#### Taking input from user[.get_attribute('value')]

```python
search = driver.find_element_by_id("twotabsearchtextbox")
searching = search.get_attribute('value')
searching
```

```python
btn = driver.find_element_by_id("nav-search-submit-button")
btn.click()
```

Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details"
and “Product URL”.

```python
#Getting page urls
product_urls = []
start=0
end=3
for page in range(start,end):#for loop for scrapping 3 page
    url=driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]')       #scraping urls
    for i in url:
        product_urls.append(i.get_attribute("href"))                        #appending the urls in product_urls list
    nxt_button=driver.find_element_by_xpath("//li[@class='a-last']//a")     #scraping the list of buttons from the page
    nxt_button.click()
    time.sleep(2)
```

```python
#Scrapping brand names
brandname = []
for i in product_urls:
    driver.get(i)
    try:
        brand = driver.find_elements_by_xpath("//span[@class='a-size-large product-title-word-break']")
        for j in brand:
            brandname.append(j.text)
    except:
        brandname.append("-")
print("Length is ",len(brandname))
brandname[:5]
```

```python
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import time
Price = []

for url in product_urls:
    driver.get(url)
    time.sleep(2)
     # Extracting price from xpath    
    try:
        price = driver.find_element_by_xpath('//td[@class="a-span12"]')  
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
print(len(Price))
```

```python
rating = []
for url in product_urls:
    driver.get(url)
    time.sleep(2)
     # Extracting ratings from xpath    
    try:
        rate = driver.find_element_by_xpath("//i[@class='a-popover-trigger a-declarative']//i//span")  
        rating.append(rate.text)
    except NoSuchElementException:
        rating.append('-')
print(len(rating))
rating[:5]
```

```python
No_of_Rating =[]
for url in product_urls:
    driver.get(url)
    time.sleep(2)
     # Extracting ratings from xpath    
    try:
        norate = driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")  
        No_of_Rating.append(norate.text)
    except NoSuchElementException:
        No_of_Rating.append('-')
print(len(No_of_Rating))
No_of_Rating[:5]
```

```python
Return = []
for url in product_urls:
    driver.get(url)
    time.sleep(2)
     # Extracting ratings from xpath    
    try:
        returns = driver.find_element_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal']")  
        Return.append(returns.text)
    except NoSuchElementException:
        Return.append('-')
print(len(Return))
Return[:5]
```

```python
Delivery = []
for url in product_urls:
    driver.get(url)
    time.sleep(2)
     # Extracting ratings from xpath    
    try:
        delivery_date = driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']//b")  
        Delivery.append(delivery_date.text)
    except NoSuchElementException:
        Delivery.append('-')
print(len(Delivery))
Delivery[:5]
```

```python
Availabitity = []
for url in product_urls:
    driver.get(url)
    time.sleep(2)
     # Extracting ratings from xpath    
    try:
        available = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-success']")  
        Availabitity.append(available.text)
    except NoSuchElementException:
        Availabitity.append('-')  
print(len(Availabitity))
```

```python
Details = []
for url in product_urls:
    driver.get(url)
    time.sleep(2)
     # Extracting ratings from xpath    
    try:
        details = driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']//li")  
        Details.append(details.text)
    except NoSuchElementException:
        Details.append('-')  
print(len(Details))
Details[:5]
```

```python
shoe = pd.DataFrame({})
shoe["Product"] = brandname
shoe["Price"] = Price
shoe["Number of ratings"] = No_of_Rating
shoe["Return"] = Return
shoe["Expected Delivery"] = Delivery
shoe["Availability"] = Availabitity
shoe["Details"] = Details
shoe["Product URL"] = product_urls
shoe
```
