Scrape the details of most viewed videos on YouTube from Wikipedia:

You need to find following details:
* Rank
* Name
* Artist
* Upload date
* Views

```python
driver = webdriver.Chrome("chromedriver.exe")
```

```python
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/"
driver.get(url)
```

```python
video_click = driver.find_element_by_xpath("//div[@class='mw-search-result-heading']//a")
video_click.click()
```

```python
#Extracting details from page
rank = []
name = []
artist = []
date = []
views = []

#scraping rank
r = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[1]")
for i in r:
    if i.text is None:
        rank.append("-")
    else:
        rank.append(i.text)
    
time.sleep(4)
#Scraping video name
n = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[2][1]")
for i in n:
    if i.text is None:
        name.append("-")
    else:
        name.append(i.text)

time.sleep(4)
#Scraping Artist
a = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[3]")
for i in a:
    if i.text is None:
        artist.append("-")
    else:
        artist.append(i.text)
    
#Scraping Date    
time.sleep(4)
d = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[5]")
for i in d:
    if i.text is None:
        date.append("-")
    else:
        date.append(i.text)

    
time.sleep(4)
v = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[4]")
for i in v:
    if i.text is None:
        views.append("-")
    else:
        views.append(i.text)
time.sleep(4)
#Creating DataFrame
videos = pd.DataFrame({
    "Rank":rank[0:30],
    "Video Name":name[0:30],
    "Artist":artist[0:30],
    "Date":date[0:30],
    "Views(in Billions)":views[0:30]
})
videos
```

Scrape the details of trending repositories on Github.com.

* Language used

```python
url = "https://github.com/"
driver.get(url)
```

```python
#Searching trending repositories
search = driver.find_element_by_xpath("//button[@class='btn-link d-lg-none mt-1 js-details-target']")
search.click()

time.sleep(3)

searching = driver.find_element_by_xpath("//nav[@class='mt-0 px-3 px-lg-0 mb-5 mb-lg-0']/ul/li[4]")
searching.click()
```

```python
#Navigating to trending page
trending = driver.find_element_by_xpath("//nav[@class='mt-0 px-3 px-lg-0 mb-5 mb-lg-0']/ul/li[4]/details/div/ul[2]/li[3]/a")
trending.click()
```

```python
title = []
description = []
count = []
language = []
page_urls = []

#Scrapping title
t = driver.find_elements_by_xpath("//article[@class='Box-row']/h1")
for i in t:
    if i.text is None:
        title.append("-")
    else:
        title.append(i.text)
titles = []
for k in range(0,len(title)):
    titles.append(title[k].split()[2])
        
time.sleep(4)
#Scrapping description
d = driver.find_elements_by_xpath("//article[@class='Box-row']/p")
for i in d:
    if i.text is None:
        description.append("-")
    else:
        description.append(i.text)
time.sleep(4)        
        
#Scrapping count
urls = driver.find_elements_by_xpath('//article[@class="Box-row"]/h1/a') 
for i in urls:
    page_urls.append(i.get_attribute('href'))
for j in page_urls:
    driver.get(j)
    time.sleep(1)
    c = driver.find_elements_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[last()-1]/div/h2/a/span")
    for i in c:
        if i.text is None:
            count.append("-")
        else:
            count.append(i.text)
time.sleep(4)        

#Scrapping languages
for j in page_urls:
    driver.get(j)
    l = driver.find_elements_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[last()]/div/ul")
    for i in l:
        if i.text is None:
            language.append("-")
        else:
            language.append(i.text.replace("\n",""))
            
print(len(titles),len(description),len(count),len(language),len(page_urls))
```

```python
trending_repo = pd.DataFrame({
    "Title":titles[:22],
    "Description":description[:22],
    "Contributors count":count[:22],
    "Language used":language[:22]
}) 
trending_repo
```
